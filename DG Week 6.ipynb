```python
import os
import pandas as pd
import math


# Get the total number of rows and columns
num_rows = df.shape[0]
num_columns = df.shape[1]

# Get the file size
file_size = os.path.getsize('D:\\AM\\Week 6\\Iowa_Liquor_Sales.csv')

# Convert file size to human-readable format
def convert_size(size_bytes):
    if size_bytes == 0:
        return "0B"
    size_names = ("B", "KB", "MB", "GB", "TB", "PB", "EB", "ZB", "YB")
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    size = round(size_bytes / p, 2)
    return f"{size}{size_names[i]}"

file_size = convert_size(file_size)

# Print the summary
print("File Summary:")
print(f"Total number of rows: {num_rows}")
print(f"Total number of columns: {num_columns}")
print(f"File size: {file_size}")
```

```python
import os
import gzip
import pandas as pd
import math

# Specify the path to the gzipped file
gz_file_path = 'D:\AM\Week 6\output_file.txt.gz'

# Read the gzipped file using gzip.open
with gzip.open(gz_file_path, 'rt') as file:
    # Read the file using pandas read_csv
    df = pd.read_csv(file, sep='|')

# Get the total number of rows and columns
num_rows = df.shape[0]
num_columns = df.shape[1]

# Get the file size
file_size = os.path.getsize(gz_file_path)

# Convert file size to human-readable format
def convert_size(size_bytes):
    if size_bytes == 0:
        return "0B"
    size_names = ("B", "KB", "MB", "GB", "TB", "PB", "EB", "ZB", "YB")
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    size = round(size_bytes / p, 2)
    return f"{size}{size_names[i]}"

file_size = convert_size(file_size)

# Print the summary
print("File Summary:")
print(f"Total number of rows: {num_rows}")
print(f"Total number of columns: {num_columns}")
print(f"File size: {file_size}")
```
